```python
import pandas as pd
df=pd.read_csv(r"train1.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.plot(df['id'], df['sales'])
plt.xlabel("id")
plt.ylabel("sales")
plt.title("sales Thrend Over Time")
plt.show()
```

```python
df['moving_avg']=df['sales'].rolling(window=3).mean()
df.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['id'], df['sales'], label='Actual Sales')
plt.plot(df['id'], df['moving_avg'], label='Moving Average')
plt.xlabel("ID")
plt.ylabel("Sales")
plt.title("Sales vs Moving Average")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x=df[['id']]     #input
y=df[['sales']] #output
```

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
plt.scatter(x, y, color='blue', label='Actual Sales')
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.xlabel("ID")
plt.ylabel("Sales")
plt.title("Linear Regression: Sales Prediction")
plt.legend()
plt.show()
```
